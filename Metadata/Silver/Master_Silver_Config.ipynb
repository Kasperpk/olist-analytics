```python
@dataclass
class Master_Silver_Schema:
    """Comprehensive schema definition for silver metadata configurations.
    
    Defines the structure for processing silver tables including the destination of the table, column name,
    data type transformations. It also includes definition of primary key and slowly changing dimensions
    columns.
    
    This schema is separate from API ingestion metadata to enable:
    - Independent evolution of transformation rules
    - Clear separation of concerns between ingestion and transformation
    - Easier maintenance by different teams
    """

    schema: StructType = None

    def __init__(self):
        """Initialize the schema by building all component schemas."""
        self.schema = self.build_schema()

    def build_schema(self):
        """Build the complete silver master schema with all nested components.
        
        :return: StructType - the complete silver metadata schema.
        """
        # -------------------------
        # FINAL MASTER SCHEMA
        # -------------------------
        silver_master_schema = StructType([
            
            # Source linkage - links back to Bronze metadata
            StructField("SourceSystem", StringType(), True),          # e.g., "Economics", "Jibble", "MicrosoftTenant"
            StructField("SourceObjectName", StringType(), True),      # e.g., "Customers" - matches bronze ObjectName
            StructField("BronzeTablePath", StringType(), True),       # e.g., "Economics_v1/Customers" - explicit path to bronze table
            
            # Silver table target
            StructField("SilverSystemName", StringType(), True),      # e.g., "economics" - schema/folder in silver
            StructField("SilverTableName", StringType(), True),       # e.g., "customers" - table name in silver

            # Transformation configuration
            StructField("SilverColumns", ArrayType(ArrayType(StringType())), True),  # [[source_name, target_name, target_type], ...]
            StructField("PK", ArrayType(StringType()), True),                        # Primary key columns
            StructField("Type2Columns", ArrayType(StringType()), True),              # SCD Type 2 tracking columns
            StructField("PartitionBy", ArrayType(StringType()), True),               # Partition columns
            StructField("OrderBy", ArrayType(StringType()), True),                   # Order by for deduplication
        ])
        
        return silver_master_schema

    def get_schema(self):
        """Return the built silver master schema.
        
        :return: StructType - the complete silver metadata schema.
        """
        return self.schema
```
