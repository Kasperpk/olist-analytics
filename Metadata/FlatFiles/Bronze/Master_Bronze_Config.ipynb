```python
@dataclass
class Master_Bronze_Schema:
    """Comprehensive schema definition for Bronze metadata configurations.
    
    Defines the structure for storing API configuration including authentication,
    pagination, URL construction, and silver layer mapping specifications.
    """

    schema: StructType = None

    def __init__(self):
        """Initialize the schema by building all component schemas."""
        self.schema = self.build_schema()

    def build_schema(self):
        """Build the complete bronze master schema with all nested components."""
        
        bronze_master_schema = StructType([
        StructField("SourceSystem", StringType(), True),
        StructField("SQLTableName", StringType(), True),         
        StructField("RootPath", StringType(), True),      
        StructField("TableName", StringType(), True),
        StructField("FileFormat", StringType(), True),])       

        return bronze_master_schema

    def get_schema(self):

        return self.schema


```
