```python
df_silver_vehicle = read_delta_table('silver.kaggle.vehicle')
```

```python
df_gold = (df_silver_vehicle.
           withColumn('Vehicle_SK', lit(0))
           .select(
               'VinNumber',
               'VehicleNumber_BK',
               'Model',
               'Manufacturer',
               'ModelYear',
               'ElectricRange',
               'VehicleType',
               col('CleanAlternativeFuelVehicleAfvEligibility').alias('CAFV'),
               'VehicleLocation'
           ))
```

```python
df_with_sk = add_surrogate_key(df_gold, 'Vehicle_SK', 'VehicleNumber_BK')
```

```python
df_dim = (df_with_sk.select(
    'Vehicle_SK',
    'VehicleNumber_BK',
    'Model',
    'Manufacturer',
    'ModelYear',
    'ElectricRange',
    'VehicleType',
    'CAFV',
    'VehicleLocation',
))
```
