```python
from pyspark.sql.functions import *
from datetime import datetime
from pyspark.sql.types import *
from dataclasses import dataclass
import re
```

```python
def read_file(file_path, file_format):
    df = spark.read.format(file_format).options(header=True).load(file_path)
    return df

def read_delta_table(table_name):
    df = spark.read.table(table_name)
    return df

def write_dataframe_to_delta_table(df, table_name, table_properties=None, write_mode='overwrite'):
    df.write.format("delta").mode(write_mode).saveAsTable(table_name)
```

```python
def clean_column_names(df):
    for col_name in df.columns:
        df = df.withColumnRenamed(col_name, expr(f"regexp_replace('{col_name}', '[^a-zA-Z0-9_]', '_')"))
    return df
```
